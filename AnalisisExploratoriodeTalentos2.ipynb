```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos hipotéticos del niño
datos = {
    "Factor de Personalidad": ["Evitativo", "Dependiente", "Obsesivo-Compulsivo", "Narcisista", "Histriónico"],
    "Puntuación": [4, 3, 2, 3, 5]  # Puntuaciones hipotéticas del niño en cada factor
}

# Crear un DataFrame de pandas
df = pd.DataFrame(datos)

# Crear un gráfico de escalones (stairs) y un gráfico de líneas (step)
plt.figure(figsize=(12, 8))

# Gráfico de escalones
factor_labels = df["Factor de Personalidad"].values
puntuaciones = df["Puntuación"].values
plt.stairs(puntuaciones, np.arange(len(factor_labels) + 1), baseline=None, fill=True, color='skyblue', alpha=0.1)

# Gráfico de líneas
plt.step(df.index, df["Puntuación"], where='mid', color='grey', linestyle='--', marker='o')
plt.xticks(df.index, factor_labels, rotation=45)

plt.xlabel("Factor de Personalidad")
plt.ylabel("Puntuación")
plt.title("Puntuación del Niño en Diferentes Factores de Personalidad")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos hipotéticos del niño
datos = {
    "Factor de Personalidad": ["Evitativo", "Dependiente", "Obsesivo-Compulsivo", "Narcisista", "Histriónico"],
    "Puntuación": [4, 3, 2, 3, 5]   # Puntuaciones hipotéticas del niño en cada factor
}

# Crear un DataFrame de pandas
df = pd.DataFrame(datos)

# Crear un gráfico de escalones no lineales
plt.figure(figsize=(10, 6))

# Preparar datos para el gráfico de escalones
puntuaciones = df["Puntuación"].values
factor_labels = df["Factor de Personalidad"].values
factor_positions = np.arange(len(factor_labels))

# Crear bordes para los escalones no lineales
bordes = np.arange(len(factor_labels) + 1)

# Crear un gráfico de escalones no lineales
plt.stairs(puntuaciones, bordes, baseline=None, fill=True, color='skyblue', alpha=0.5)

# Gráfico de líneas
plt.plot(factor_positions, puntuaciones, 'o--', color='grey', alpha=0.7)

plt.xlabel("Factor de Personalidad")
plt.ylabel("Puntuación")
plt.title("Puntuación del Niño en Diferentes Factores de Personalidad")
plt.xticks(factor_positions, factor_labels, rotation=45)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del niño (por ejemplo, puntuaciones en cinco dimensiones de personalidad)
dimensiones = ['Evitativo', 'Dependiente', 'Obsesivo-Compulsivo', 'Narcisista', 'Histriónico', 'Antisocial', 'Esquizoide', 'Límite', 'Esquizotípico']
# Puntuaciones para un niño (por ejemplo)
padreFamilia = [1, 5, 3, 0, 4, 3, 2, 4, 5]
ninio = [3, 2, 4, 5, 2, 4.6, 5, 3.8, 3]
nuestro = [1, 3, 4, 3, 4, 5, 5, 3.8, 3]

dimensiones2 = ['Extraversión', 'Amabilidad', 'Conciencia', 'Estabilidad Emocional', 'Apertura a la Experiencia']
padreFamilia2 = [5, 4.2, 4.0, 3.8, 3.6]
ninio2 = [2, 4.6, 5, 3.8, 3]
nuestro2 = [4, 5, 5, 3.8, 3]

# Crear los bins y los centros para el Factor 1
bins_factor1 = np.arange(len(dimensiones) + 1)
centers_factor1 = bins_factor1[:-1]

# Crear los bins y los centros para el Factor 2
bins_factor2 = np.arange(len(dimensiones2) + 1)
centers_factor2 = bins_factor2[:-1]

# Datos para la función stairs() - Factor 1
y_factor1 = np.array(padreFamilia)
# Datos para el reporte alterno - Factor 1
y_nuestro_factor1 = np.array(nuestro)
# Datos para el reporte alterno - Factor 1
y_alterno_factor1 = np.array(ninio)

# Datos para la función stairs() - Factor 2
y_factor2 = np.array(padreFamilia2)
# Datos para el reporte alterno - Factor 2
y_nuestro_factor2 = np.array(nuestro2)
# Datos para el reporte alterno - Factor 2
y_alterno_factor2 = np.array(ninio2)

# Crear subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Puntuaciones de Personalidad del Niño')

# Subplot 1: Puntuaciones de Personalidad - Evaluación Padre/Familia (Factor 1)
axs[0, 0].step(centers_factor1, y_factor1, 'o--', where='post', label='Evaluación Padre/Familia', color='gray')
axs[0, 0].set_title('Puntuaciones de Personalidad - Padre/Familia (Factor 1)')
axs[0, 0].set_xticks(centers_factor1)
axs[0, 0].set_xticklabels(dimensiones, rotation=90)
axs[0, 0].grid(axis='y', linestyle='--', alpha=0.5)
axs[0, 0].legend()

# Subplot 2: Puntuaciones de Personalidad - Evaluación Niño (Factor 1)
axs[0, 1].stairs(y_alterno_factor1, bins_factor1, baseline=None, label='Evaluación Niño', color='blue', alpha=0.4)
axs[0, 1].set_title('Puntuaciones de Personalidad - Niño (Factor 1)')
axs[0, 1].set_xticks(centers_factor1)
axs[0, 1].set_xticklabels(dimensiones, rotation=90)
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.5)
axs[0, 1].legend()

# Subplot 3: Puntuaciones de Personalidad - Evaluación Nuestra (Factor 1)
axs[0, 2].stairs(y_nuestro_factor1, bins_factor1, baseline=None, label='Nuestra Evaluación', color='red', alpha=0.4)
axs[0, 2].set_title('Puntuaciones de Personalidad - Nuestra Evaluación (Factor 1)')
axs[0, 2].set_xticks(centers_factor1)
axs[0, 2].set_xticklabels(dimensiones, rotation=90)
axs[0, 2].grid(axis='y', linestyle='--', alpha=0.5)
axs[0, 2].legend()

# Subplot 4: Puntuaciones de Personalidad - Evaluación Padre/Familia (Factor 2)
axs[1, 0].step(centers_factor2, y_factor2, 'o--', where='post', label='Evaluación Padre/Familia', color='gray')
axs[1, 0].set_title('Puntuaciones de Personalidad - Padre/Familia (Factor 2)')
axs[1, 0].set_xticks(centers_factor2)
axs[1, 0].set_xticklabels(dimensiones2, rotation=90)
axs[1, 0].grid(axis='y', linestyle='--', alpha=0.5)
axs[1, 0].legend()

# Subplot 5: Puntuaciones de Personalidad - Evaluación Niño (Factor 2)
axs[1, 1].stairs(y_alterno_factor2, bins_factor2, baseline=None, label='Evaluación Niño', color='blue', alpha=0.4)
axs[1, 1].set_title('Puntuaciones de Personalidad - Niño (Factor 2)')
axs[1, 1].set_xticks(centers_factor2)
axs[1, 1].set_xticklabels(dimensiones2, rotation=90)
axs[1, 1].grid(axis='y', linestyle='--', alpha=0.5)
axs[1, 1].legend()

# Subplot 6: Puntuaciones de Personalidad - Evaluación Nuestra (Factor 2)
axs[1, 2].stairs(y_nuestro_factor2, bins_factor2, baseline=None, label='Nuestra Evaluación', color='red', alpha=0.4)
axs[1, 2].set_title('Puntuaciones de Personalidad - Nuestra Evaluación (Factor 2)')
axs[1, 2].set_xticks(centers_factor2)
axs[1, 2].set_xticklabels(dimensiones2, rotation=90)
axs[1, 2].grid(axis='y', linestyle='--', alpha=0.5)
axs[1, 2].legend()

# Ajustar el espacio entre subplots para que las etiquetas no se solapen
plt.tight_layout()
plt.subplots_adjust(top=0.85)

# Mostrar la gráfica
plt.show()
```
