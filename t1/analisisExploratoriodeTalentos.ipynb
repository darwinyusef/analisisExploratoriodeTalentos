```python
import matplotlib.pyplot as plt
import numpy as np

category_names = ['Bajo', 'Medio', 'Alto']
results = {
    'Semana 2: Deportes': [42, 8, 58],
    'Semana 3: Idiomas': [36, 25, 47],
    'Semana 4: Lectura y Escritura': [39, 7, 62],
    'Semana 5: Teatro y Actuación': [43, 26, 39],
    'Semana 6: Club de Debate y Oratoria': [50, 5, 53],
    'Semana 7: Matemáticas y Ajedrez': [27, 35, 46],
    'Semana 8: Naturaleza y Animales': [38, 32, 38],
    'Semana 9: Ética y Religión': [29, 19, 60]
}

def survey(results, category_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5, label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncols=len(category_names), bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')

    return fig, ax

survey(results, category_names)
plt.xlabel('Talentos')
plt.ylabel('Talentos explorados')
#plt.title('Talentos Específicos en Niveles Alto, Medio y Bajo')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres de las categorías de inteligencia
categorias = [
    'Lógico-Matemática', 'Estructural-Lingüística', 'Interpersonal',
    'Intrapersonal', 'Cinestésica', 'Armónica-Musical',
    'Espacial-Visual', 'Naturalista', 'Existencial', 'Creativa'
]

# Calificación promedio en cada categoría (valores de 0 a 10)
calificaciones = [8, 9, 7, 8, 6, 5, 7, 9, 6, 7]

# Agregar la primera categoría al final para cerrar el gráfico
categorias.append(categorias[0])
calificaciones.append(calificaciones[0])

# Ángulos para las categorías
angulos = np.linspace(0, 2 * np.pi, len(categorias), endpoint=False)

# Ajustar el ángulo para que el gráfico comience desde la parte superior
angulos = np.concatenate((angulos, [angulos[0]]))

# Calificación promedio para cerrar el gráfico
calificaciones = np.concatenate((calificaciones, [calificaciones[0]]))

# Crear la figura y el eje polar
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

# Dibujar el gráfico de radar
ax.fill(angulos, calificaciones, 'b', alpha=0.3)

# Establecer los nombres de las categorías como etiquetas en los ángulos
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(categorias)
ax.set_yticklabels([])  # Ocultar etiquetas en el radio

# Establecer el título
plt.title('Calificación Promedio en Categorías de Inteligencia')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de las inteligencias múltiples y sus niveles de desarrollo
inteligencias = np.array([
    [9, 9, 9, 9, 9, 9, 9, 9, 9], #
    [0, 8, 7, 6, 9, 7, 9, 8, 9],
    [7, 6, 7, 6, 8, 8, 8, 8, 7],
    [2, 4, 1, 0, 2, 3, 4, 1, 3],
     [1, 2, 9, 0, 2, 3, 4, 1, 3],
])



# Colormap
cmap = plt.get_cmap("Reds")

# Crear la figura
fig, ax = plt.subplots(figsize=(12, 8))
cax = ax.imshow(inteligencias, cmap=cmap, aspect='auto', vmin=0, vmax=9)

# Etiquetas y título
plt.xlabel('Tipos de Inteligencia')
plt.ylabel('Niveles de Desarrollo')
plt.title('Calificación de Niveles de Desarrollo de Inteligencias Múltiples bajo Factores de personalidad (Big Five)')

# Barra de color
cbar = fig.colorbar(cax)

# Función format_coord personalizada
def format_coord(x, y):
    col = round(x)
    row = round(y)
    nrows, ncols = inteligencias.shape
    nivel = ""
    if 0 <= col < ncols and 0 <= row < nrows:
        z = inteligencias[row, col]
        if z <= 2:
            nivel = "Extraversión"
        elif z <= 4:
            nivel = "Amabilidad"
        elif z <= 6:
            nivel = "Conciencia"
        elif z <= 8:
            nivel = "Apertura a la experiencia"
        else:
            nivel = "Estabilidad emocional"
        return f'Inteligencia: {row+1}, Nivel: {nivel}'
    else:
        return f'x={x:1.4f}, y={y:1.4f}'

ax.format_coord = format_coord

# Etiquetas de ejes
inteligencia_labels = [
    'Lógico-Matemática', 'Estructural-Lingüística', 'Interpersonal',
    'Intrapersonal', 'Cinestésica', 'Armónica-Musical',
    'Espacial-Visual', 'Naturalista', 'Existencial'
]

plt.xticks(np.arange(inteligencias.shape[1]), inteligencia_labels, rotation=45)
niveles_labels = ['Extraversión', 'Amabilidad', 'Conciencia', 'Apertura a la experiencia', 'Estabilidad emocional']
plt.yticks(np.arange(len(niveles_labels)), niveles_labels)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Datos de las inteligencias múltiples y sus niveles de desarrollo
inteligencias = np.array([
    [9, 9, 9, 9, 9, 9, 9, 9, 9],
    [0, 8, 7, 6, 9, 7, 9, 8, 9],
    [7, 6, 7, 6, 8, 8, 8, 8, 7],
    [2, 4, 1, 0, 2, 3, 4, 1, 3],
    [1, 2, 9, 0, 2, 3, 4, 1, 3],
])

# Crear la figura y ajustar el estilo de Seaborn
sns.set_theme()
plt.figure(figsize=(12, 10))

# Colormap
cmap = sns.color_palette("PuOr", as_cmap=True)

# Crear el heatmap con Seaborn
ax = sns.heatmap(inteligencias, cmap=cmap, annot=True, fmt="d", linewidths=0.5, cbar=True, vmin=0, vmax=9)

# Etiquetas y título
ax.set_xlabel('Tipos de Inteligencia')
ax.set_ylabel('Niveles de Desarrollo')
plt.title('Calificación de Niveles de Desarrollo de Inteligencias Múltiples bajo Factores de personalidad (Big Five)')

# Etiquetas de ejes
inteligencia_labels = [
    'Lógico-Matemática', 'Estructural-Lingüística', 'Interpersonal',
    'Intrapersonal', 'Cinestésica', 'Armónica-Musical',
    'Espacial-Visual', 'Naturalista', 'Existencial'
]

ax.set_xticklabels(inteligencia_labels, rotation=90)
niveles_labels = ['Extraversión', 'Amabilidad', 'Conciencia', 'Apertura a la experiencia', 'Estabilidad emocional']
ax.set_yticklabels(niveles_labels, rotation=0)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
